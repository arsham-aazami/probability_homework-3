```python
import matplotlib.pyplot as plt
import math

def calculate_bionmial_pmf_and_mean(n, p):
    # S(X) = {0, 1, 2, ...}
    # Create a list of possible outcomes and their probabilities
    probailities = [math.comb(n, k)*( (1-p)**(n-k) )*(p**k) for k in range(n+1)]
    # Calculate the mean of the random variable X
    mean = 0
    for k in range(n+1) : mean += k*probailities[k]
    # Plot the probability mass function
    fig, ax = plt.subplots(figsize=(10, 7))
    ax.stem(probailities)
    plt.xlabel(f"X ~ Binomial(n={n}, p={p})")
    plt.ylabel(f"P(X = x)")
    plt.show()
    print(f"mean of the random variable X = {mean}")
```

```python
calculate_bionmial_pmf_and_mean(50, 0.7)
```

```python

import matplotlib.pyplot as plt


def calculate__geometric_pmf_and_mean(n, p):
    #S(Y) = {1, 2, 3, ...}
    # Create a list of possible outcomes and their probabilities
    pmfs = [(1-p)**(k-1)*p for k in range(0, n+1)]
    # Calculate the expected value of the random variable X
    expected_value = sum(pmfs)
    # Plot the probability mass function
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.stem(pmfs)
    plt.xlabel(f"Y ~ Geometric(n={n}, p={p})")
    plt.ylabel(f"pmf")
    plt.show()
    print(f"mean of the random variable Y = {expected_value}")
```

```python
calculate__geometric_pmf_and_mean(50, 0.7)
```

```python
import matplotlib.pyplot as plt
import random
import math

def calculate_bionmial_pmf_and_mean(n, p):
    # S(X) = {0, 1, 2, ...}
    # Create a list of possible outcomes and their probabilities
    events = []
    binary_list = [0, 1]
    for i in range(10):
        result = random.choice(binary_list)
        events.append(result)
    print(events)

    # 0 represent failure
    # 1 represent success
    probailities = [math.comb(n, k)*( (1-p)**(n-k) )*(p**k) for k in range(n+1)]
    # Calculate the mean of the random variable X
    mean = n*p
    # Plot the probability mass function
    fig, ax = plt.subplots(figsize=(10, 7))
    ax.stem(probailities)
    plt.xlabel(f"X ~ Binomial(n={n}, p={p})")
    plt.ylabel(f"P(X = x)")
    plt.show()
    print(f"mean of the random variable X = {mean}")
```

```python
calculate_bionmial_pmf_and_mean(50, 0.7)
```

```python

import matplotlib.pyplot as plt

def calculate_geometric_pmf_and_mean(n, p):
    # S(Y) = {1, 2, 3, ...}
    # Create a list of possible outcomes and their probabilities
    probailities = [(1 - p)**(k - 1) * p for k in range(1, n + 1)]
    # Calculate the expected value of the random variable X
    mean = 1 / p
    # Plot the probability mass function
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.stem(range(1, n + 1), probailities)
    
    plt.xlabel(f"Y ~ Geometric(n={n}, p={p})")
    plt.ylabel("pmf")
    plt.title("Geometric Probability Mass Function")
    plt.show()
    
    print(f"mean of the random variable Y = {mean}")
```

```python
calculate_geometric_pmf_and_mean(50, 0.7)
```

```python
import random
import matplotlib.pyplot as plt

def geometric_distribution(p, n):
    """Simulates the geometric distribution using relative frequency."""
    count = 0
    for i in range(n):
        if random.random() < p:
            count += 1
    return count

# Example usage
p = 0.7
n = 50
results = [geometric_distribution(p, n) for _ in range(10000)]
relative_frequency = sum(results) / (n * len(results))

# Plot the PMF
x = range(1, n + 1)
pmf = [results.count(i) / len(results) for i in x]
plt.stem(x, pmf, use_line_collection=True)
plt.title(f"Geometric Distribution PMF with p={p} and n={n}")
plt.xlabel("Number of trials needed to get the first success")
plt.ylabel("Probability")
plt.show()

```

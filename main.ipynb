```python
import random
import matplotlib.pyplot as plt

def binomial_distribution(p, n):
    """Simulates the binomial distribution using relative frequency."""
    count = 0
    for i in range(n):
        if random.random() < p:
            count += 1
    return count

# Example usage
p = 0.9
n = 500
results = [binomial_distribution(p, n) for _ in range(10000)]
relative_frequency = sum(results) / (n * len(results))

# Plot the PMF
x = range(1, n + 1)
pmf = [results.count(i) / len(results) for i in x]
mean = n*p 
plt.stem(x, pmf, use_line_collection=True)
plt.title(f"Binomial Distribution PMF with p={p} and n={n}")
plt.xlabel("Number of trials needed to get the first success")
plt.ylabel("Probability")
plt.show()
print(f"The Expected value of random variable X is {mean}")


```

```python
import random
import matplotlib.pyplot as plt
from collections import Counter

def geometric_distribution(p, num_simulations):
    """simulate a geometric distribution"""
    results = []
    for _ in range(num_simulations):
        trials_until_success = 1
        while random.random() > p:
            trials_until_success += 1
        results.append(trials_until_success)
    return results

def plot_geometric_pmf(results):
    counts = Counter(results)
    max_trials = max(results)

    plt.stem(counts.keys(), [count / len(results) for count in counts.values()], use_line_collection=True)
    plt.title('Geometric Distribution PMF (Simulation)')
    plt.xlabel('Number of Trials until Success')
    plt.ylabel('Probability')
    plt.xticks(range(1, max_trials + 1))
    plt.show()

n = 50
p = 0.7
mean = 1/p
results = geometric_distribution(p, n)
plot_geometric_pmf(results)
print(mean)

```
